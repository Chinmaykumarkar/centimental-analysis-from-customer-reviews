```python
#Removing punctuations

puncs="!”#$%&'()*+,-./:;?@[\]^_`{|}~"
for i in df["reviewText"]:
    newstr=""
    for j in i:
        if j in puncs:
            newstr+=""
        else:
            newstr+=j
    df.loc[df['reviewText']==i,'reviewText']=newstr
```

```python
df
```

```python
#Converting to lowercase

for i in df["reviewText"]:
    df.loc[df['reviewText']==i,'reviewText']=i.lower()
```

```python
df
```

```python
#Tokenization

def tokenize(a):
    lst=a.split()
    return lst

df["reviewText"]=df["reviewText"].apply(lambda x: tokenize(x))
```

```python
df
```

```python
#Removing garbage words

import nltk
nltk.download('stopwords')

garbagewords = nltk.corpus.stopwords.words('english')
type(garbagewords)
```

```python
def removegw(l):
    lst=list(set(l)-set(garbagewords))
    return lst

df["reviewText"]=df["reviewText"].apply(lambda x: removegw(x))
```

```python
df
```

```python
#Lemmatization - reducing the words to their base form without losing the meaning

Lemmatizer = nltk.stem.WordNetLemmatizer()

def lemmating(a):
    newa=[]
    for i in a:
        newa.append(Lemmatizer.lemmatize(i))
    return newa

df['reviewText']=df['reviewText'].apply(lambda x: lemmating(x))
```

```python
df
```

```python
#Embedding and padding

s=0
for i in df["reviewText"]:
    if len(i)>s:
        s=len(i)
s
```

```python
from keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences
```

```python
tokenizer = Tokenizer(num_words=500, split=' ')
X = tokenizer.fit_on_texts(df['reviewText'].values)
X = tokenizer.texts_to_sequences(df['reviewText'].values)
```

```python
X = pad_sequences(X)
```

```python
X
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,df["overall"],test_size=0.30)

model1=MLPClassifier(hidden_layer_sizes=(500,500,500),activation='relu',solver='adam',max_iter=100)
model1.fit(X_train,y_train)

predict_train=model1.predict(X_train)
predict_test=model1.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,predict_test))
print(classification_report(y_test,predict_test))
```

```python
from sklearn.naive_bayes import MultinomialNB
model2 = MultinomialNB()
model2.fit(X_train,y_train)

predict_train=model2.predict(X_train)
predict_test=model2.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,predict_test))
print(classification_report(y_test,predict_test))
```
